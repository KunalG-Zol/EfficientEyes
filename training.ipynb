# Creating a validation set from within the training data

```python
import os
import shutil
import random

# --- CONFIGURATION ---
DATASET_ROOT = "./dataset"
IMAGES_ROOT = os.path.join(DATASET_ROOT, "images")
LABELS_ROOT = os.path.join(DATASET_ROOT, "labels")

# Define Paths
TRAIN_IMG = os.path.join(IMAGES_ROOT, "train")
TRAIN_LBL = os.path.join(LABELS_ROOT, "train")

VAL_IMG = os.path.join(IMAGES_ROOT, "val")
VAL_LBL = os.path.join(LABELS_ROOT, "val")

TEST_IMG = os.path.join(IMAGES_ROOT, "test")
TEST_LBL = os.path.join(LABELS_ROOT, "test")

# Split Ratio (20% of Train goes to Val)
VAL_SPLIT = 0.20

def move_files(file_list, src_img_dir, dst_img_dir, src_lbl_dir, dst_lbl_dir):
    count = 0
    total = len(file_list)
    
    os.makedirs(dst_img_dir, exist_ok=True)
    os.makedirs(dst_lbl_dir, exist_ok=True)
    
    print(f"Moving {total} images to {dst_img_dir}...")
    
    for filename in file_list:
        # 1. Move Image
        shutil.move(os.path.join(src_img_dir, filename), 
                    os.path.join(dst_img_dir, filename))
        
        # 2. Move Label (Change .png/.jpg to .txt)
        label_name = os.path.splitext(filename)[0] + ".txt"
        lbl_src = os.path.join(src_lbl_dir, label_name)
        lbl_dst = os.path.join(dst_lbl_dir, label_name)
        
        if os.path.exists(lbl_src):
            shutil.move(lbl_src, lbl_dst)
            
        count += 1
        if count % 1000 == 0:
            print(f"  - Moved {count}/{total} files...")
            
    print(f"Finished moving {count} files.")


# 1. Safety Check: If 'test' exists, we probably already ran this.
if os.path.exists(TEST_IMG):
    print("'dataset/images/test' already exists. Aborting to prevent double-splitting.")
else:
    # 2. Rename current 'val' (Original Test Set) to 'test'
    print(f"Renaming current 'val' folder to 'test' (Preserving Official Test Set)...")
    os.rename(VAL_IMG, TEST_IMG)
    os.rename(VAL_LBL, TEST_LBL)
    
    # 3. Create new empty 'val' folders
    os.makedirs(VAL_IMG, exist_ok=True)
    os.makedirs(VAL_LBL, exist_ok=True)

    # 4. Get list of training images
    all_images = [f for f in os.listdir(TRAIN_IMG) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    random.shuffle(all_images)
    
    # 5. Calculate split
    num_val = int(len(all_images) * VAL_SPLIT)
    val_files = all_images[:num_val]
    
    print(f"Found {len(all_images)} training images.")
    print(f"Moving {num_val} images (20%) to new Validation set...")
    
    # 6. Execute Move
    move_files(val_files, TRAIN_IMG, VAL_IMG, TRAIN_LBL, VAL_LBL)
    
    print("\nDataset Split Complete!")
    print(f"  Train: {len(os.listdir(TRAIN_IMG))} images")
    print(f"  Val:   {len(os.listdir(VAL_IMG))} images (Subset of Train)")
    print(f"  Test:  {len(os.listdir(TEST_IMG))} images (Original RarePlanes Test)")
```

```python
import yaml
import os

data_yaml = {
    'path': os.path.abspath('dataset'),
    'train': 'images/train',
    'val':   'images/val',
    'test':  'images/test',  
    'names': {
        0: 'aircraft'
    }
}

with open('dataset/data_split.yaml', 'w') as f:
    yaml.dump(data_yaml, f, default_flow_style=False)

print("Created dataset/data_split.yaml")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8l.pt') 

results = model.train(
    data='./dataset/data_split.yaml',
    epochs=100, 
    imgsz=640, 
    batch=128,
    device=0,
    project='rareplanes_run',
    name='yolov8_large_split_v1',
    exist_ok=False,
    verbose=True,
    cache=True
)
```
